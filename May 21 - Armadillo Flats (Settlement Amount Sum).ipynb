```python
def settlement_sum(i):
    df = pd.read_csv('ArmadilloFlats_May21.csv')
    df2 = df.iloc[0:1000]
    df3 = df.iloc[5576:]
    df4 = df.iloc[1000:5576]
    df5 = df4.copy()
    df4 = df4.sample(frac=1)
    df4.reset_index(inplace=True, drop=True)
    df4['Volume'] = df5['Volume']
    df6 = pd.concat([df2, df4, df3], axis =0)
    df6['Fixed Minus Float'] = df6['Fixed Price'] - df6['Index Price']
    df6['Settlement Amount'] = df6['Volume'] * df6['Fixed Minus Float']
    return df6['Settlement Amount'].sum(axis = 0, skipna = True)
```

```python
for i in range(1000):
    sum = settlement_sum(i)
    print (sum)
```
